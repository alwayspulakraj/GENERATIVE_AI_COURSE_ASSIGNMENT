#1.What is the difference between static and dynamic variables in Python.

Dynamic variables are defined inside a function or a method and a new instance is created each time the function is called.

```python
#example-
class MyClass:
    static_variable = 10
    def function(self, age):
        if age > 50:
            dynamic_variable = 20
        else:
            dynamic_variable = 30
        print("Static Variable:", MyClass.static_variable)
        print("Dynamic Variable:", dynamic_variable)
        return dynamic_variable
```

#23.How does inheritance work in python? Explain all types of inheritance with an example.
Inheritance is a concept in object-oriented programming that allows one class to inherit the properties and behavior of another class.



```python
#example-(single inheritance)
class Animal:
    def sound(self):
        print("The animal makes a sound.")
class Dog(Animal):
    def sound(self):
        print("The dog barks.")
my_dog = Dog()
my_dog.sound()
```

```python
#example-(multiple inheritance)
class Bird:
    def fly(self):
        print("I can fly!")
class Mammal:
    def feed(self):
        print("I can feed my young.")
class Bat(Bird, Mammal):
    pass
my_bat = Bat()
my_bat.fly()
my_bat.feed()
```

```python
#example-(multilevel inheritance)
class Vehicle:
    def move(self):
        print("I can move.")
class Car(Vehicle):
    def drive(self):
        print("I can drive.")
class ElectricCar(Car):
    def charge(self):
        print("I can charge my battery.")
my_car = ElectricCar()
my_car.move()
my_car.drive()
my_car.charge()
```

```python
#example- (Hierarchical Inheritance)
class Shape:
    def area(self):
        pass
class Circle(Shape):
    def area(self):
        print("The area of the circle is pi * r^2.")
class Rectangle(Shape):
    def area(self):
        print("The area of the rectangle is length * width.")
my_circle = Circle()
my_circle.area()
my_rectangle = Rectangle()
my_rectangle.area()
```

```python
#example-(Hybrid Inheritance)
class Animal:
    def sound(self):
        print("The animal makes a sound.")
class Mammal:
    def feed(self):
        print("I can feed my young.")
class Dog(Animal, Mammal):
    pass
class Cat(Animal):
    pass
my_dog = Dog()
my_dog.sound()
my_dog.feed()
my_cat = Cat()
my_cat.sound()
```

```python
#2. Design and implement a Python program for managing student information using object-oriented
#principles. Create a class called `Student` with encapsulated attributes for name, age, and roll number.
#Implement getter and setter methods for these attributes. Additionally, provide methods to display student
#information and update student details.
Tasks:
# Define the `Student` class with encapsulated attributes.
# Implement getter and setter methods for the attributes
# Write methods to display student information and update details.
# Create instances of the `Student` class and test the implemented functionality.
```

```python
class Student:
    def __init__(self, age, roll_number,name=""):
        self.__name = name
        self.__age = age
        self.__roll_number = roll_number
    @property
    def get_name(self):
        return self.__name
    @property
    def get_age(self):
        return self.__age
    @property
    def get_roll_number(self):
        return self.__roll_number
    @get_name.setter
    def set_name(self, name):
        self.__name = name
    @get_age.setter
    def set_age(self, age):
        self.__age = age
    @get_roll_number.setter
    def set_roll_number(self, roll_number):
        self.__roll_number = roll_number
    # Method to display student info
    def display_info(self):
        print(f"Name-{self.__name}")
        print(f"Age-{self.__age}")
        print(f"Roll Number- {self.__roll_number}")
    # Method to update student details
    def update_details(self, age=None, roll_number=None ,name=""):
        if name:
            self.__name = name
        if age:
            self.__age = age
        if roll_number:
            self.__roll_number = roll_number
#objects of the Student class
student1 = Student(name="Pulakraj", age=20, roll_number=1)
student2 = Student(name="narendra modi", age=22, roll_number=2)
#Display
print("Student 1 Information:")
student1.display_info()
print("\nStudent 2 Information:")
student2.display_info()
#update
student1.update_details(name="Subrat", age=21)
student2.update_details(roll_number=3)
print("\nUpdated Student 1 Information:")
student1.display_info()
print("\nUpdated Student 2 Information:")
student2.display_info()
```

#35.A random sample of size 25 from a population gives the sample standard derivation to be 9.0. Test the hypothesis that the population standard derivation is 10.5. Hint(Use chi-square distribution).

```python
#35.A random sample of size 25 from a population gives the sample standard derivation to be 9.0. Test the hypothesis that the population standard derivation is 10.5. Hint(Use chi-square distribution).
import scipy.stats as stats
n=25
s=9.0
sigma=10.5
chi2 = (n-1)*(s**2/sigma**2)
print("Chi-square statistic-", chi2)
nu=n-1
print("Degree of freedom-", nu)
p_value = stats.chi2.sf(chi2, nu)
print("P-value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("The population standard deviation is not 10.5")
else:
    print("The population standard deviation is 10.5")
```
